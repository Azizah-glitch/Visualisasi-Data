```python
#Plot garis dengan judul dan  label
import matplotlib.pyplot as plt
#Data
x=[52,93,15,72,61,21,76,22,11,75]
y=[108,156,151,140,137,160,198,192,166,112]
#Membuat plot garis
plt.plot(x,y)
#Menambahkan judul dan label
plt.title("Plot Garis")
plt.xlabel("Sumbu X")
plt.ylabel("Sumbu Y")
#Menampilkan plot
plt.show
```

```python
#Scatter plot
import matplotlib.pyplot as plt
#Data
x=[52,93,15,72,61,21,76,22,11,75]
y=[108,156,151,140,137,160,198,192,166,112]
#Membuat scatter plot
plt.scatter(x,y)
#Menambahkan judul dan label
plt.title("Scatter Plot")
plt.xlabel("Sumbu X")
plt.ylabel("Sumbu Y")
#Menampilkan plot
plt.show
```

```python
#Scatter plot dengan Trendline
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.array([52,93,15,72,61,21,76,22,11,75])
y = np.array([108,156,151,140,137,160,198,192,166,112])

# Buat scatter plot
plt.scatter(x, y, color="hotpink", marker="o", label="Data")

# Hitung trendline (regresi linear)
m, b = np.polyfit(x, y, 1)   # 1 berarti linear
trendline = m * x + b

# Gambar trendline
plt.plot(x, trendline, color="red", linewidth=2, label=f"Trendline (y={m:.2f}x+{b:.2f})")

# Tambahkan judul dan label
plt.title("Scatter Plot dengan Trendline")
plt.xlabel("Sumbu X")
plt.ylabel("Sumbu Y")
plt.legend()
plt.grid(alpha=0.3)

# Tampilkan plot
plt.show()
```

```python
#Bar plot
import matplotlib.pyplot as plt
#Data
x=[52,93,15,72,61,21,76,22,11,75]
y=[108,156,151,140,137,160,198,192,166,112]
#Membuat bar plot
plt.bar(x, y)
#Menambahkan judul dan label
plt.title("Bar Plot")
plt.xlabel("Sumbu X")
plt.ylabel("Sumbu Y")
#Menampilkan plot
plt.show
```

```python
#Histogram
import matplotlib.pyplot as plt
#Data
x=[52,93,15,72,61,21,76,22,11,75]

#Membuat Histogram
plt.hist(x, bins=5, color="pink", edgecolor="black")
#Menambahkan judul dan label
plt.title("Histogram Data X")
plt.xlabel("Nilai")
plt.ylabel("Frekuensi")
#Menampilkan plot
plt.show
```

```python
#Pie Chart
import matplotlib.pyplot as plt

# Data
penjualan = [40, 25, 20, 15]   # nilai (satu array angka)
produk = ["A", "B", "C", "D"]  # label (opsional, kategori)

# Buat pie chart
plt.pie(penjualan, labels=produk, autopct='%1.1f%%', colors=["pink","skyblue","lightgreen","orange"])
plt.title("Distribusi Penjualan Produk")
plt.show()
```

```python
#Heatmap
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Buat data random
np.random.seed(42)
data = pd.DataFrame(np.random.randint(1,100,(6,6)),
                    columns=list("ABCDEF"))

# Buat heatmap
plt.figure(figsize=(8,6))
sns.heatmap(data, annot=True, cmap="YlGnBu", fmt="d")  # annot=True = tampilkan angka
plt.title("Heatmap")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Baca data
file_path = r"C:\Users\LENOVO\OneDrive\Desktop\Visualisasi data\Data_Regresi.csv"
df = pd.read_csv(file_path)

# Bersihkan nama kolom
df.columns = df.columns.str.strip()

# Buat figure
plt.figure(figsize=(10,6))

# Buat boxplot untuk tiap kolom numerik
sns.boxplot(data=df[["Promo(X1)", "Karyawan(X2)", "Jual(Y)"]], palette="Set2")

plt.title("Boxplot Data Penjualan")
plt.xlabel("Variabel")
plt.ylabel("Nilai")
plt.grid(alpha=0.3)
plt.show()
```
